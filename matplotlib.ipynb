## markers 

 https://www.w3schools.com/python/matplotlib_markers.asp
 
 colors:['r','g','b','c','m','y','k','w']

```python
for symbol in ['o','*','.','+','P','s','|','_']:
    plt.plot(ypoints, marker = symbol , linestyle="")
    plt.show()
```

## fmt
fmt : marker|line|color

https://www.w3schools.com/python/matplotlib_markers.asp

```python
plt.plot(ypoints, 'o--r') # just y sent 
plt.show()
```

```python
for line_style in ['-',":","--","-."]:
    #markerfacecolor or mfc
    #markersize or ms
    #markeredgecolor or mec
    
    plt.plot(ypoints, f'o{line_style}c',ms = 15,mfc = 'w',mec="r") 
    plt.show()
```

ms = marker size 
mec = marker edge color
mfc = marker face color 

can use hexadecimal color => forexample : mfc = '#4CAF50'
can use color_names in https://www.w3schools.com/colors/colors_names.asp => forexample : mfc = 'hotpink'

```python
# ColorMap
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors,cmap='viridis')
plt.colorbar()

plt.show()


plt.scatter(x, y, c=colors,cmap='Accent_r')
plt.colorbar()

plt.show()
```

# Pie Charts

```python
y = np.array([35, 25, 25, 15]) #  start axis is 0 degree axis
plt.pie(y)
plt.show() 

y = np.array([1, 2, 3, 4]) # 10
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y,labels = mylabels, startangle = 90,colors = mycolors)
plt.legend(title = "Four Fruits:")

plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.3,2, 0, 0.1]
plt.pie(y, labels = mylabels, explode = myexplode,shadow = True)

plt.show() 
```
