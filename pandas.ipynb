# series

```python
a = [1, 7, 2]

myvar = pd.Series(a) # like array , like column of table or a row
print(myvar)
print(myvar[0])
```

```python
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
print(myvar["x"])
print(myvar[0])
```

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

# DataFrame

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45 ]
}

df = pd.DataFrame(data)

print(df) 
df
```

# loc and iloc

```python

df = pd.DataFrame([
    [1,2],
    [2,3],
    [4,5]
]) # can use index , columns 
print(df)
print("__________")
print(df.loc[0]) 
print(df.loc[0][0])

df.index = ["day1", "day2", "day3"]
df.columns=["calories","duration"]
print(df) 

print("__________")
print(df.loc["day1"]) # can't use df.loc[0]
print(df.loc["day1"]["calories"]) # can't use df.loc["day1"][0]
```

```python
print(df)
print("_________________")
print(df.iloc[0])
print("_________________")
print(df.iloc[0][0])
print(df.iloc[0,0])
print(df.iloc[:2 , :1]) # all numpy array subscriptes
```

# agg

```python

data = {
  "x": [50, 40, 30],
  "y": [300, 1112, 42]
}

df = pd.DataFrame(data)

x = df.agg(["sum" , "std" , "mean" ,"count"]) # axis

print(x)
```

# append

```python

data1 = {
  "age": [16, 14, 10],
  "qualified": [True, True, True]
}
df1 = pd.DataFrame(data1)
print(df1)
print("______________")

data2 = {
  "age": [55, 40],
  "qualified": [True, False]
}
df2 = pd.DataFrame(data2)

newdf = df1.append(df2)

print(newdf)
```
